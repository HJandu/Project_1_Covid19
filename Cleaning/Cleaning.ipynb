```python
#importing pandas
import pandas as pd
```

```python
#defining the file path
csv_path = "../CSVs/indicators-CountiesUAsfromApr2021.data.csv"
```

```python
#converting to dataframe
df = pd.read_csv(csv_path)
df.T
```

```python
#getting rid of non-suicide data
non_suicide = []

for x in range(len(df)):
    if df["Indicator Name"][x] != "Suicide rate":
        non_suicide.append(x)
```

```python
#creating a dataframe from that
cleaned_df = df.drop(non_suicide)
cleaned_df.reset_index(drop=True,inplace = True)
cleaned_df.T
```

```python
#getting rid of data from before the 2017-19 period
not_app = []

for x in range(len(cleaned_df)):
    if cleaned_df["Time period"][x] != "2019 - 21" and cleaned_df["Time period"][x] != "2017 - 19":
        not_app.append(x)
```

```python
#turning that into a dataframe
applicable_df = cleaned_df.drop(not_app)
applicable_df.reset_index(drop=True, inplace = True)
applicable_df.T
```

```python
#splitting the data into pre and during covid
pre_covid = []

for x in range(len(applicable_df)):
    if applicable_df["Time period"][x] == "2017 - 19":
        pre_covid.append(x)

covid = []

for x in range(len(applicable_df)):
    if applicable_df["Time period"][x] == "2019 - 21":
        covid.append(x)
```

```python
#creating dataframes from that
covid_times = applicable_df.drop(pre_covid)
pre_covid_times = applicable_df.drop(covid)
```

```python
#resetting the index
covid_times.reset_index(drop=True,inplace = True)
pre_covid_times.reset_index(drop=True,inplace = True)
pre_covid_times.T
```

```python
covid_times.T
```

```python
#a list for the columns that are going to be dropped
drop_col = ["Indicator ID", "Indicator Name", "Parent Code", "Area Code", "Category Type", "Age","Lower CI 95.0 limit", "Upper CI 95.0 limit", "Lower CI 99.8 limit", "Upper CI 99.8 limit", "Denominator", "Value note", "Recent Trend", "Time period Sortable", "New data", "Compared to goal", "Time period range"]
```

```python
#dropping the columns
pre_covid_times.drop(columns = drop_col, inplace = True)
covid_times.drop(columns = drop_col, inplace = True)
```

```python
#getting rid of data not needed for specific questions (regions and areas) and making dataframes from that
not_region = []

for x in range(len(covid_times)):
    if covid_times["Area Type"][x] != "Region" or covid_times["Sex"][x] != "Persons":
        not_region.append(x)

not_area = []

for x in range(len(covid_times)):
    if covid_times["Area Type"][x] != "Counties & UAs (from Apr 2021)" or covid_times["Sex"][x] != "Persons":
        not_area.append(x)

covid_regions = covid_times.drop(not_region)
covid_areas = covid_times.drop(not_area)
covid_regions.reset_index(drop=True, inplace=True)
covid_areas.reset_index(drop=True, inplace=True)
covid_regions.drop(columns = ["Sex", "Category", "Parent Name", "Area Type", "Compared to Region value or percentiles"], inplace = True)
covid_areas.drop(columns = ["Sex", "Category", "Area Type"], inplace = True)

not_region = []

for x in range(len(pre_covid_times)):
    if pre_covid_times["Area Type"][x] != "Region" or pre_covid_times["Sex"][x] != "Persons":
        not_region.append(x)

not_area = []

for x in range(len(pre_covid_times)):
    if pre_covid_times["Area Type"][x] != "Counties & UAs (from Apr 2021)" or pre_covid_times["Sex"][x] != "Persons":
        not_area.append(x)

pre_covid_regions = pre_covid_times.drop(not_region)
pre_covid_areas = pre_covid_times.drop(not_area)
pre_covid_regions.reset_index(drop=True, inplace=True)
pre_covid_areas.reset_index(drop=True, inplace=True)
pre_covid_regions.drop(columns = ["Sex", "Category", "Parent Name", "Area Type", "Compared to Region value or percentiles"], inplace = True)
pre_covid_areas.drop(columns = ["Sex", "Category", "Area Type"], inplace = True)
```

```python
#getting rid of data not needed for specific question (gender) and making dataframes from that
not_gender = []

for x in range(len(pre_covid_times)):
    if pre_covid_times["Area Type"][x] != "England" or pd.isna(pre_covid_times["Category"][x]) == False or pre_covid_times["Sex"][x] == "Persons":
        not_gender.append(x)

pre_covid_gender = pre_covid_times.drop(not_gender)
pre_covid_gender.reset_index(drop=True, inplace=True)
pre_covid_gender.drop(columns = ["Parent Name", "Area Name", "Area Type", "Category", "Compared to England value or percentiles", "Compared to Region value or percentiles"], inplace=True)

not_gender = []

for x in range(len(covid_times)):
    if covid_times["Area Type"][x] != "England" or pd.isna(covid_times["Category"][x]) == False or covid_times["Sex"][x] == "Persons":
        not_gender.append(x)

covid_gender = covid_times.drop(not_gender)
covid_gender.reset_index(drop=True, inplace=True)
covid_gender.drop(columns = ["Parent Name", "Area Name", "Area Type", "Category", "Compared to England value or percentiles", "Compared to Region value or percentiles"], inplace=True)
```

```python
#getting rid of data not needed for specific question (gender) and making dataframes from that
not_wealth = []

for x in range(len(pre_covid_times)):
    if pre_covid_times["Area Type"][x] != "England" or pd.isna(pre_covid_times["Category"][x]) == True or pre_covid_times["Sex"][x] == "Male" or pre_covid_times["Sex"][x] == "Female":
        not_wealth.append(x)

pre_covid_wealth = pre_covid_times.drop(not_wealth)
pre_covid_wealth.reset_index(drop=True, inplace=True)
pre_covid_wealth.drop(columns = ["Parent Name", "Area Name", "Area Type", "Sex", "Compared to Region value or percentiles"], inplace=True)

not_wealth = []

for x in range(len(covid_times)):
    if covid_times["Area Type"][x] != "England" or pd.isna(covid_times["Category"][x]) == True or covid_times["Sex"][x] == "Male" or pre_covid_times["Sex"][x] == "Female":
        not_wealth.append(x)

covid_wealth = covid_times.drop(not_wealth)
covid_wealth.reset_index(drop=True, inplace=True)
covid_wealth.drop(columns = ["Parent Name", "Area Name", "Area Type", "Sex", "Compared to Region value or percentiles"], inplace=True)
```
