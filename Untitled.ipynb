```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
covid_regions=pd.read_csv("DuringCovid_r.csv")
covid_areas=pd.read_csv("DuringCovid_a.csv")
pre_covid_regions=pd.read_csv("PreCovid_r.csv")
pre_covid_areas=pd.read_csv("PreCovid_a.csv")
```

```python
covid_regions
```

```python
not_persons = []
for x in range(len(covid_regions)):
    if covid_regions["Sex"][x] != "Persons":
        not_persons.append(x)

covid_regions.drop(not_persons, inplace=True)
covid_regions.reset_index(drop=True, inplace=True)
covid_regions.drop(columns="Sex", inplace=True)

not_persons = []
for x in range(len(pre_covid_regions)):
    if pre_covid_regions["Sex"][x] != "Persons":
        not_persons.append(x)

pre_covid_regions.drop(not_persons, inplace=True)
pre_covid_regions.reset_index(drop=True, inplace=True)
pre_covid_regions.drop(columns="Sex", inplace=True)
```

```python
x_axis = np.arange(len(covid_regions["Area Name"]))

plt.bar(x_axis - 0.35/2, pre_covid_regions["Count"], 0.35, label='Pre Covid')
plt.bar(x_axis + 0.35/2, covid_regions["Count"], 0.35, label='Covid')

regions = ["NE", "NW", "Y", "EM", "WM", "E", "L", "SE", "SW"]

plt.xticks(x_axis, regions)
plt.legend()

plt.show;
```

```python
x_axis = np.arange(len(covid_regions["Area Name"]))

plt.bar(x_axis - 0.35/2, pre_covid_regions["Value"], 0.35, label='Pre Covid')
plt.bar(x_axis + 0.35/2, covid_regions["Value"], 0.35, label='Covid')

regions = ["NE", "NW", "Y", "EM", "WM", "E", "L", "SE", "SW"]

plt.xticks(x_axis, regions)
plt.legend()

plt.show;
```
