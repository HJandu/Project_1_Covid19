```python
x_axis = np.arange(len(pre_covid["Category"]))
  
plt.bar(x_axis - 0.2, pre_covid["Value"], 0.4, label = "Pre-Covid")
plt.bar(x_axis + 0.2, during_covid["Value"], 0.4, label = "During Covid")
  
plt.xticks(x_axis, pre_covid["Category"])
plt.xlabel("Decile of Deprived Area")
plt.ylabel("Rate of Suicide per 100,000 People")
plt.ylim(8)
plt.legend()
#plt.savefig("Output/RateComparisonWealth")
plt.show()
```
