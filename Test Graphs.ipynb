```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
during_covid_csv = "DuringCovid.csv"
pre_covid_csv = "PreCovid.csv"
```

```python
during_covid = pd.read_csv(during_covid_csv)
pre_covid = pd.read_csv(pre_covid_csv)
```

```python
during_covid.head()
```

```python
during_covid.info()
```

```python
pre_covid.head()
```

```python
counts=pre_covid.groupby(["Sex"])["Value"].count()
counts.plot.pie(title="Male vs Female Suicide rates",label="",autopct="%1.1f%%")
explode = (0.1, 0.1, 0)
```

```python
piePlot=pre_covid.groupby(["Sex"])["Value"].count().reset_index(name="Value")
labs = piePlot["Sex"]
Count = piePlot["Value"]
explode = (0.1, 0.1, 0)
colors = ["pink","teal", "yellow"]
plt.pie(Count, explode=explode, labels=labs, colors=colors,autopct="%1.1f%%")
plt.title("Suicide Rates by Gender pre Covid")

plt.show()
```

```python
during_covid.groupby(["Sex"])["Value"].count().plot.pie(title="Male vs Female Suicide rates",label="",autopct="%1.1f%%")
```

```python
piePlot=during_covid.groupby(["Sex"])["Value"].count().reset_index(name="Value")
labs = piePlot["Sex"]
Count = piePlot["Value"]
explode = (0.1, 0.1, 0)
colors = ["pink","teal", "yellow"]
plt.pie(Count, explode=explode, labels=labs, colors=colors,autopct="%1.1f%%")
plt.title("Suicide Rates by Gender pre Covid")

plt.show()
```
