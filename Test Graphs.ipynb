```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
```

```python
during_covid_csv = "DuringCovid.csv"
pre_covid_csv = "PreCovid.csv"
```

```python
during_covid = pd.read_csv(during_covid_csv)
pre_covid = pd.read_csv(pre_covid_csv)
```

```python
during_covid.head()
```

```python
during_covid.info()
```

```python
pre_covid.head()
```

```python
pre_covid.drop(pre_covid.loc[pre_covid['Sex']=="Persons"].index, inplace=True)
```

```python
piePlot=pre_covid.groupby(["Sex"])["Count"].sum().reset_index(name="Count")
labs = piePlot["Sex"]
Count = piePlot["Count"]
explode = (0.1, 0.1)
colors = ["pink","teal"]
plt.pie(Count, explode=explode, labels=labs, colors=colors,autopct="%1.1f%%")
plt.title("Suicide Rates by Gender Pre Covid")

plt.show()
```

```python
during_covid.drop(during_covid.loc[during_covid['Sex']=="Persons"].index, inplace=True)
```

```python
piePlot=during_covid.groupby(["Sex"])["Count"].sum().reset_index(name="Count")
labs = piePlot["Sex"]
Count = piePlot["Count"]
explode = (0.1, 0.1)
colors = ["pink","teal"]
plt.pie(Count, explode=explode, labels=labs, colors=colors,autopct="%1.1f%%")
plt.title("Suicide Rates by Gender During Covid")

plt.show()
```

```python
summary_area_name=pre_covid.groupby(["Area Type"])
```

```python
summary_area_name.head()
```

```python
region_df = pre_covid.loc[(pre_covid["Area Type"] == "Region")]
region_df.head()
```

```python
g = sns.catplot(x = "Area Name",       # x variable name
            y = "Value",       # y variable name
            hue = "Sex",  # group variable name
            data = region_df,     # dataframe to plot
            kind = "bar",
           errorbar= None,
            legend='auto'
               )
ax = g.facet_axis(0, 0)
for p in ax.patches:
    ax.text(p.get_x() + 0.015, 
            p.get_height() * 1.02, 
            '{0:.1f}'.format(p.get_height()), 
            color='black', rotation='horizontal', size='medium')
plt.xticks(rotation=90)
plt.title("Gender comparison by area Pre Covid")
plt.ylim(0, 25)
plt.show()
```
