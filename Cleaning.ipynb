```python
#importing pandas
import pandas as pd
```

```python
#defining the file path
csv_path = "indicators-CountiesUAsfromApr2021.data.csv"
```

```python
#converting to dataframe
df = pd.read_csv(csv_path)
df.T
```

```python
#getting rid of non-suicide data
non_suicide = []

for x in range(len(df)):
    if df["Indicator Name"][x] != "Suicide rate":
        non_suicide.append(x)
```

```python
#creating a dataframe from that
cleaned_df = df.drop(non_suicide)
cleaned_df.reset_index(inplace = True)
cleaned_df
```

```python
#getting rid of data from before the 2017-19 period
not_app = []

for x in range(len(cleaned_df)):
    if cleaned_df["Time period"][x] != "2019 - 21" and cleaned_df["Time period"][x] != "2017 - 19":
        not_app.append(x)
```

```python
#turning that into a dataframe
applicable_df = cleaned_df.drop(not_app)
applicable_df.reset_index(inplace = True)
applicable_df.T
```

```python
#splitting the data into pre and during covid
pre_covid = []

for x in range(len(applicable_df)):
    if applicable_df["Time period"][x] == "2017 - 19":
        pre_covid.append(x)

covid = []

for x in range(len(applicable_df)):
    if applicable_df["Time period"][x] == "2019 - 21":
        covid.append(x)
```

```python
#creating dataframes from that
covid_times = applicable_df.drop(pre_covid)
pre_covid_times = applicable_df.drop(covid)
```

```python
#resetting the index
covid_times.drop(columns = "level_0", inplace = True)
pre_covid_times.drop(columns = "level_0", inplace = True)

covid_times.reset_index(inplace = True)
pre_covid_times.reset_index(inplace = True)
pre_covid_times.T
```

```python
covid_times.T
```

```python
#a list for the columns that are going to be dropped
drop_col = ["level_0", "index", "Indicator ID", "Indicator Name", "Parent Code", "Area Code", "Category Type", "Age","Lower CI 95.0 limit", "Upper CI 95.0 limit", "Lower CI 99.8 limit", "Upper CI 99.8 limit", "Denominator", "Value note", "Recent Trend", "Time period Sortable", "New data", "Compared to goal", "Time period range"]
```

```python
#dropping the columns
pre_covid_times.drop(columns = drop_col, inplace = True)
covid_times.drop(columns = drop_col, inplace = True)
```
